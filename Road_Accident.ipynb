```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
#load the dataset
df = pd.read_csv('accident.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
# Visualize the distribution of accident severity
plt.figure(figsize=(10, 6))
sns.countplot(x='Accident_severity', data=df)
plt.title('Distribution of Accident Severity')
plt.show()
```

```python
# Analyze the impact of weather conditions on accidents
plt.figure(figsize=(12, 6))
sns.countplot(x='Weather_conditions', hue='Accident_severity', data=df)
plt.title('Impact of Weather Conditions on Accident Severity')
plt.xticks(rotation=90)
plt.show()
```

```python
# Analyze the impact of road surface type on accidents
plt.figure(figsize=(12, 6))
sns.countplot(x='Road_surface_type', hue='Accident_severity', data=df)
plt.title('Impact of Road Surface Type on Accident Severity')
plt.xticks(rotation=90)
plt.show()
```

```python
# Analyze the impact of light conditions on accidents
plt.figure(figsize=(12, 6))
sns.countplot(x='Light_conditions', hue='Accident_severity', data=df)
plt.title('Impact of Light Conditions on Accident Severity')
plt.xticks(rotation=90)
plt.show()
```

```python
# Analyze the impact of driver's age band on accident severity
plt.figure(figsize=(12, 6))
sns.countplot(x='Age_band_of_driver', hue='Accident_severity', data=df)
plt.title('Impact of Driver\'s Age Band on Accident Severity')
plt.xticks(rotation=90)
plt.show()
```

```python
# Analyze the impact of driving experience on accident severity
plt.figure(figsize=(12, 6))
sns.countplot(x='Driving_experience', hue='Accident_severity', data=df)
plt.title('Impact of Driving Experience on Accident Severity')
plt.xticks(rotation=90)
plt.show()
```

Conclusion:-
- Most of the accidents happen in the normal weather conditions maybe drivers are careless during normal days,
- Most of the accidents happen in the asphalt roads,
- Most of the accidents happen in the day time,
- Drivers of age 18-30 with experience of 5-10 years are mostly involved in accidents.
